```python
import krakenex
import datetime
import calendar
import pandas as pd
import time
import os.path
import numpy as np
import requests

%matplotlib inline
```

```python
period = 1.05

def check_trades(c = 'ETH', relay = 'PLN', since = 0):
    url = 'https://bitbay.net/API/Public/' + c + relay + '/trades.json?since=' + str(since)
    time.sleep(period)
    r = requests.get(url)
#     print(r.json())
    return r.json()
```

```python
if os.path.isfile('Data/trades_BitBay_ETH.csv'):
    df_trades = pd.read_csv('Data/trades_BitBay_ETH.csv', index_col=0, header=0)
else:
    df_trades = pd.DataFrame(columns=['date','price','type','amount','tid'])

#print(df_trades.dtypes)
    
if os.path.isfile('Data/last_BitBay_ETH.txt'):
    file = open('Data/last_BitBay_ETH.txt', 'r') 
    last = int(file.read())
    file.close()
else:
    last = 0

while True:
# for i in range(1):
    try:
        trades = check_trades(since = last)
    except:
        print("except")
        time.sleep(period*10)
        pass
    else:
#         if not trades['error']:
        if len(trades) > 0:
            last = trades[-1]['tid']

            df_temp = pd.DataFrame(data=trades, columns=['date','price','type','amount','tid'])
            df_temp[['amount','price']] = df_temp[['amount','price']].astype('float64')

            df_trades = pd.concat([df_trades,df_temp], ignore_index=True).drop_duplicates().sort_values('date').reset_index(drop=True)  

            df_trades.to_csv('Data/trades_BitBay_ETH.csv')
            file = open('Data/last_BitBay_ETH.txt', 'w')
            file.write(last)
            file.close()
            
            file = open('Data/last_check_ETH.txt', 'w')
            file.write(str(time.time()))
            file.close()

            print(pd.to_datetime(df_temp['date'], unit='s').tail(1))
            time.sleep(10*period)
        else:
            time.sleep(10*period)
#         else:
#             print(trades['error'])
#             time.sleep(period*10)
```
