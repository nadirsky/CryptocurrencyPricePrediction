```python
import krakenex
import datetime
import calendar
import pandas as pd
import time
import os.path
import numpy as np

%matplotlib inline
```

```python
if os.path.isfile('trades.csv'):
    df_trades = pd.read_csv('trades.csv', index_col=0, header=0)
else:
    df_trades = pd.DataFrame(columns=['timestamp','volume','price','buysell','marketlimit','misc'])

#print(df_trades.dtypes)
    
if os.path.isfile('last.txt'):
    file = open('last.txt', 'r') 
    last = int(file.read())
    file.close()
else:
    last = 0

while True:
#for i in range(1):
    try:
        k = krakenex.API()
        k.load_key('kraken.key')
        trades = k.query_public('Trades', {'pair': 'XXBTZUSD', 'since': last})
    except:
        print("except")
        time.sleep(150)
        pass
    else:
        if not trades['error']:
            last = trades['result']['last']
            btc_trades = np.array(trades['result']['XXBTZUSD'])
            btc_trades[:,0], btc_trades[:,2] = btc_trades[:,2], btc_trades[:,0].copy()
            #print(btc_trades)
            df_temp = pd.DataFrame(data=btc_trades, columns=['timestamp','volume','price','buysell','marketlimit','misc'])
            df_temp[['timestamp','volume','price']] = df_temp[['timestamp','volume','price']].astype('float64')
            #print(df_temp.head(1))
            print(pd.to_datetime(df_temp['timestamp'], unit='s').tail(1))
            print(df_temp.tail(1))
            #print(df_temp.dtypes)

            df_trades = pd.concat([df_trades,df_temp], ignore_index=True).drop_duplicates().sort_values('timestamp').reset_index(drop=True)  

            df_trades.to_csv('trades.csv')
            file = open('last.txt', 'w')
            file.write(last)
            file.close()

            time.sleep(150)
        else:
            print(trades['error'])
            time.sleep(150)

#print(df_trades)

#df_spread.plot(x='timestamp', y=['price','volume'])
```
